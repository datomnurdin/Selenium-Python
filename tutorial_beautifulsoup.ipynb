```python
from bs4 import BeautifulSoup
import requests
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
response= requests.get("https://www.ebay.com/sch/i.html?_from=R40&_trksid=m570.l1313&_nkw=charizard+gx+hyper+rare&_sacat=0", headers=headers)
```

```python
response.status_code
```

```python
response.content
```

```python
soup = BeautifulSoup(response.text,"html.parser")
print(soup.prettify())
```

```python
soup_link = soup.findAll("li", attrs={"class":'s-item'})
print(soup_link)
```

```python
soup_link = soup.findAll("li", attrs={"class":'s-item'})[0]
print(soup_link)
```

```python
soup_link = soup.findAll("li", attrs={"id":'srp-river-results-listing1'})[0]
print(soup_link)
```

```python
soup_link.find("h3", attrs={"class":'s-item__title'}).text
```

```python
soup_link = soup.findAll("li", attrs={"class":'s-item'})

for x in range(len(soup_link)):
    if x != 0:
        print(soup_link[x].find("h3", attrs={"class":'s-item__title'}).text)
```
